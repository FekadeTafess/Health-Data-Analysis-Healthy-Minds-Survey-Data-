```python
import pandas as pd
df = pd.read_csv("HMS_2024-2025_PUBLIC_instchars.cleaned.csv")
df.head()
```

```python
variables = [
    'deprawsc', 'anx_score', 'dep_maj', 'anx_any', 'dep_any',
    'pellgrant', 'gender_male', 'gender_female', 'age',
    'finpast', 'fincur', 'food_worry', 'housing_worry'
]
```

```python
print("--- Missing Values Before Cleaning ---")
print(df[variables].isnull().sum())
```

```python
import numpy as np
df['gender_clean'] = np.where(df['gender_male'] == 1, 1,
                             np.where(df['gender_female'] == 1, 2, np.nan))
```

```python
df['pellgrant_clean'] = df['pellgrant'].replace(2, np.nan)
```

```python
df = df.drop_duplicates(subset=['responseid'])
```

```python
df_final = df.dropna(subset=['dep_any', 'anx_any', 'fincur', 'food_worry', 'housing_worry']).copy()
```

```python
print(f"\nTotal records after removing duplicates and missing values: {len(df_final)}")
print("\n--- Summary of Key Variables ---")
print(df_final[['deprawsc', 'anx_score', 'age']].describe())

df_final.to_csv('cleaned_college_mental_health.csv', index=False)
```
