```python
import pandas as pd
import numpy as np

df = pd.read_csv('HMS_2024-2025_PUBLIC_instchars.cleaned.csv')

# 2. Select ONLY the variables needed for the research question
research_vars = ['fincur', 'food_worry', 'housing_worry', 'dep_any', 'anx_any', 'age', 'gender_male', 'gender_female']
df_subset = df[research_vars].copy()

df_subset['gender_male'] = df_subset['gender_male'].fillna(0)
df_subset['gender_female'] = df_subset['gender_female'].fillna(0)


df_subset['gender_clean'] = 3
df_subset.loc[df_subset['gender_male'] == 1, 'gender_clean'] = 1
df_subset.loc[df_subset['gender_female'] == 1, 'gender_clean'] = 2

core_research_cols = ['fincur', 'dep_any', 'anx_any']
df_final = df_subset.dropna(subset=core_research_cols).copy()

summary_table = df_final.describe()

print("--- Final Data Quality Check ---")
print(f"Total records after cleaning: {len(df_final)}")
print(f"Missing values remaining: {df_final.isnull().sum().sum()}")
print("\n--- Summary Statistics Table ---")
print(summary_table)

df_final.to_csv('final_research_dataset.csv', index=False)
summary_table.to_csv('summary_statistics.csv')

print(f"Total Rows Remaining: {len(df_final)}")
print(f"Total Missing Values: {df_final.isnull().sum().sum()}")


df_final = df_final.dropna()

print(f"Final Row Count: {len(df_final)}")
print(f"Final Missing Values: {df_final.isnull().sum().sum()}")
```
