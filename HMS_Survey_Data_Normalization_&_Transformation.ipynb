```python
import pandas as pd
import numpy as np

df = pd.read_csv('HMS_2024-2025_PUBLIC_instchars.cleaned.csv')

core_vars = ['fincur', 'food_worry', 'housing_worry', 'deprawsc', 'anx_score']

df_norm = df.dropna(subset=core_vars).copy()

for col in core_vars:
    col_min = df_norm[col].min()
    col_max = df_norm[col].max()
    df_norm[f'{col}_minmax'] = (df_norm[col] - col_min) / (col_max - col_min)

for col in core_vars:
    col_mean = df_norm[col].mean()
    col_std = df_norm[col].std()
    df_norm[f'{col}_zscore'] = (df_norm[col] - col_mean) / col_std

df_norm.to_csv('normalized_research_data.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('HMS_2024-2025_PUBLIC_instchars.cleaned.csv')

df['deprawsc_log'] = np.log1p(df['deprawsc'])

df['financial_hardship_score'] = df['food_worry'] + df['housing_worry']

scaler = StandardScaler()
cols_to_scale = ['age', 'deprawsc', 'anx_score', 'financial_hardship_score']
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

print("Transformations complete. Data is ready for Logistic Regression.")
```

```python
summary_table = df[['age', 'fincur', 'food_worry', 'housing_worry', 'deprawsc', 'anx_score']].describe()

summary_table.to_csv('summary_table.csv')
print("Summary table saved successfully.")
```
